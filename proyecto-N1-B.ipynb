```python
from selenium import webdriver
import time
import requests
```

```python

listas_id={}

categorias=['catalogos','comics-historietas','libros','manga','revistas','otros']
for i in range(len(categorias)):
    driver=webdriver.Chrome('chromedriver.exe')
    driver.get('https://libros.mercadolibre.com.ar/{}/'.format(categorias[i]))
    nombre=driver.find_element_by_id("categorySearch").get_attribute('value')
    nombre=nombre[3:]
    listas_id[categorias[i]]=nombre
    driver.close()
    
```

```python
para_pd={'categoria':[],'item':[],'precio':[]}
```

```python
para_pd
```

```python

for i in range(len(categorias)):
    
    url="http://api.mercadolibre.com/sites/MLA/search?category={}".format(listas_id[categorias[i]])
    
    args={'offset':'0'}
    count=0
    print(categorias[i].upper())
    respond=requests.get(url,params=args)
    jason=respond.json()
    jsnrslt=jason['paging']
    veces=int(jsnrslt['total'])//50
    if veces > 100:
        veces=100
    else:
        veces=int(jsnrslt['total'])//50
    for g in range(veces):
        response=requests.get(url,params=args)


        if response.status_code == 200:
            json=response.json()
            for m in json['results']:
                count+=1
                print(count,m['title'],m['price'])
                para_pd['categoria'].append(categorias[i])
                para_pd['item'].append(m['title'])
                para_pd['precio'].append(m['price'])
        args['offset']=int(args['offset'])+50
        requests.get(url,params=args)
```

```python
para_pd
```

```python
import pandas as pd
import re
```

```python
data=pd.DataFrame(para_pd)
```

```python
data['item']=data['item'].apply(lambda x: re.sub("[^a-zA-Z0-9\s]",' ',x))
```
